### Tennessee

```python
tn = pd.read_csv("Tennessee_rare_aquatic_animals_county.csv", encoding='latin1')

tn = tn.replace('--',np.NaN)

tn = tn[tn["Type"].str.contains("Vascular Plant")==False]

tn
```

```python
tn_unique = tn.drop_duplicates(subset=['Scientific Name'])
tn_unique.groupby(['Category', 'State Rank']).agg('count')
```

### Alabama

```python
al = pd.read_csv("alabama/track_2022.csv")
```

```python
# Keep if CLASS is "Crayfish and Shrimp", "Fishes", "Freshwater Mussels" or "Freshwater Snails"
al = al[al['CLASS'].isin(["Crayfish and Shrimp", "Fishes", "Freshwater Mussels", "Freshwater Snails"])]
al
```
