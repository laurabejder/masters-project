```python
import pandas as pd
import numpy as np
```

```python
df_2001 = pd.read_csv('track_2001.csv')
df_2022 = pd.read_csv('track_2022.csv')

df_2001
```

```python
df_2001.groupby('CLASS').count()
```

```python
df = pd.merge(df_2001,df_2022, on=['CLASS','SCIENTIFIC NAME'], how='outer')

df = df[df['CLASS'] == "Freshwater Snails"]
df.head(50)

df = df.dropna(subset=['YEAR_x', 'YEAR_y'])

df.head(50)
```

```python
# define the mapping of state rank values to numeric values
global_rank_mapping = {
    'GX': 0, 'G1': 1, 'G2': 2, 'G3': 3, 'G4': 4, 'G5': 5,
    'G?': np.nan, 'G1G2': 1.5, 'G2G3': 2.5, 'G3G4': 3.5, 'G4G5': 4.5,
    'GU': np.nan, 'GH': 0.5
}

# loop through the columns STATE_RANK_x and STATE_RANK_y and create new columns state_rank_x_num and state_rank_y_num
for col in ['GLOBAL RANK_x', 'GLOBAL RANK_y']:
    new_col = col.lower() + '_num'  # create the new column name
    df[new_col] = df[col].map(global_rank_mapping)  # map the state rank values to numeric values

df = df.dropna(subset=['global rank_x_num', 'global rank_y_num'])

data1 = df.loc[:, ['SCIENTIFIC NAME', 'COMMON NAME_x', 'global rank_x_num', 'global rank_y_num']]

data1['change'] = data1.apply(lambda x: 1 if x['global rank_x_num'] > x['global rank_y_num'] else (-1 if x['global rank_x_num'] < x['global rank_y_num'] else 0), axis=1)

data1
```

```python
# define the mapping of state rank values to numeric values
state_rank_mapping = {
    'SX': 0, 'S1': 1, 'S2': 2, 'S3': 3, 'S4': 4, 'S5': 5,
    'S?': np.nan, 'S1S2': 1.5, 'S2S3': 2.5, 'S3S4': 3.5, 'S4S5': 4.5,
    'SU': np.nan, 'SH': 0.5
}

# loop through the columns STATE_RANK_x and STATE_RANK_y and create new columns state_rank_x_num and state_rank_y_num
for col in ['STATE RANK_x', 'STATE RANK_y']:
    new_col = col.lower() + '_num'  # create the new column name
    df[new_col] = df[col].map(state_rank_mapping)  # map the state rank values to numeric values

df = df.dropna(subset=['state rank_x_num', 'state rank_y_num'])

data = df.loc[:, ['SCIENTIFIC NAME', 'COMMON NAME_x', 'state rank_x_num', 'state rank_y_num']]

data['change'] = data.apply(lambda x: 1 if x['state rank_x_num'] > x['state rank_y_num'] else (-1 if x['state rank_x_num'] < x['state rank_y_num'] else 0), axis=1)

data
```
